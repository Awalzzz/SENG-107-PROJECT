## Data Analysis

```python
data.quality.value_counts()
```

```python
story = data.groupby('quality').mean()
story
```

```python
data.plot(x='quality', y='volatile acidity', kind = 'hist', color='red')
plt.xlabel('Quality')
plt.ylabel('Volatile Acidity')
```

```python
data.hist(figsize=(10,10),bins=50, color='red')
plt.show()
```

```python
# Evaluating the relationship between each input variable.
plt.figure(figsize=(20,10))
sns.heatmap(data.corr(), annot=True,cmap='Reds')
plt.show()
```

### As can be observed from the correlation heat map above, there is a strong relationship between:
* Fixed acidity and Citric acid    
* Total sulphur dioxide and Free sulphur dioxide
* Alcohol and Quality



## Alcohol vs Wine Quality

```python
sns.lineplot(y='alcohol', x='quality', data=data, color='red')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
```

#### My interpretation of the graph above is that high quality rated wines contain higher alcohol content.
